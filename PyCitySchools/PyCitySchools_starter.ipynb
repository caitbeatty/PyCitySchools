## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#school_data_complete
```

```python
school_data_complete_df = pd.DataFrame(school_data_complete)
```

```python
#total # of schools
school_total = len(school_data_complete["school_name"].unique())
school_total
```

```python
#total # of students 
student_total = len(school_data_complete["Student ID"].unique())
student_total
```

```python
#total budget
budget = (school_data['budget'].sum())
budget 
```

```python
#avg math score
avg_math = (student_data["math_score"].mean())
avg_math
```

```python
#avg reading score
avg_reading = (student_data['reading_score'].mean())
avg_reading
```

```python
# % passing math 
school_data_complete["math_pass"] = school_data_complete["math_score"] >= 70
percent_math_pass = ((school_data_complete["math_pass"]).mean())*100
percent_math_pass
```

```python
# % passing reading
school_data_complete["reading_pass"] = school_data_complete["reading_score"] >= 70
percent_reading_pass = ((school_data_complete["reading_pass"]).mean())*100
percent_reading_pass
```

```python
# overall passing 
school_data_complete["overall_pass"] = (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
percent_overall_pass = ((school_data_complete["overall_pass"]).mean())*100
percent_overall_pass
```

```python
#create table
district_summary = [{"Total schools":school_total, "Total Student":student_total, "Total Budget": budget, "Average Math Score": avg_math, "Average Reading Score": avg_reading, "%Passing Math": percent_math_pass, "%Passing Reading": percent_reading_pass, "%Overall Passing": percent_overall_pass}]
district_summary_table = pd.DataFrame(district_summary)
#district_summary_table
```

```python
#format
district_summary_table['Total Budget'] = district_summary_table['Total Budget'].map("${:,.2f}".format)
district_summary_table['Total Student'] = district_summary_table['Total Student'].map("{:,}".format)
district_summary_table
```
